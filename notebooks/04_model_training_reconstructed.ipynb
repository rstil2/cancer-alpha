This notebook reconstructs the model training pipeline and regenerates all manuscript figures using the latest processed data.

```python
# Import Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import balanced_accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
import lightgbm as lgb
import xgboost as xgb
from sklearn.linear_model import LogisticRegression
import shap
import os
```

```python
# Load Processed Data
X = pd.read_csv('../models/X_processed.csv')
y = pd.read_csv('../models/y_processed.csv')['cancer_type']
print(f'Loaded X shape: {X.shape}, y shape: {y.shape}')
```

```python
# SMOTE Class Balancing
smote = SMOTE(random_state=42)
X_bal, y_bal = smote.fit_resample(X, y)
print(f'After SMOTE: X shape: {X_bal.shape}, y shape: {y_bal.shape}')
```

```python
# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X_bal, y_bal, test_size=0.2, random_state=42, stratify=y_bal)
print(f'Train shape: {X_train.shape}, Test shape: {X_test.shape}')
```

```python
# Model Training and Evaluation
models = {
    'LightGBM': lgb.LGBMClassifier(random_state=42),
    'XGBoost': xgb.XGBClassifier(random_state=42),
    'RandomForest': RandomForestClassifier(random_state=42),
    'LogisticRegression': LogisticRegression(max_iter=1000, random_state=42)
}
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    bal_acc = balanced_accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted')
    rec = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    cm = confusion_matrix(y_test, y_pred)
    results[name] = {'bal_acc': bal_acc, 'prec': prec, 'rec': rec, 'f1': f1, 'cm': cm}
    print(f'{name}: Balanced Accuracy={bal_acc:.4f}, Precision={prec:.4f}, Recall={rec:.4f}, F1={f1:.4f}')
```

```python
# Plot and Save Figures
fig_dir = '../science/figures'
os.makedirs(fig_dir, exist_ok=True)
for name, res in results.items():
    plt.figure(figsize=(8,6))
    sns.heatmap(res['cm'], annot=True, fmt='d', cmap='Blues')
    plt.title(f'{name} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.savefig(f'{fig_dir}/figure_{name.lower()}_confusion_matrix.png', dpi=300)
    plt.close()
```

```python
# SHAP Interpretability for LightGBM
explainer = shap.Explainer(models['LightGBM'], X_train)
shap_values = explainer(X_test)
plt.figure(figsize=(12,8))
shap.summary_plot(shap_values.values, X_test, plot_type='bar', show=False)
plt.title('LightGBM SHAP Feature Importance')
plt.savefig(f'{fig_dir}/figure_lightgbm_shap_importance.png', dpi=300)
plt.close()
```
